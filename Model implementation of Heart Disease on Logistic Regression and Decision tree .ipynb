```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
y = df['HeartDisease']

```

```python
x = df[['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS','RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope']]
```

```python
x_train, x_test, y_train ,y_test = train_test_split(x,y , random_state=25, train_size=0.8, shuffle=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python


li=['ChestPainType','RestingECG','ST_Slope','Sex','ExerciseAngina']
def makeordinal(li,df1):
    oe = OrdinalEncoder()
    for col in li:
        #oe.fit(x_train)
        
        df1[col]=oe.fit_transform(df1[[col]])
    return df1
```

```python
x_train=makeordinal(li,x_train)
x_train.sample(20)
```

```python
x_test = makeordinal(li,x_test)
x_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg_model = LogisticRegression()
lg_model.fit(x_train, y_train)
```

```python
y_predict = lg_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Before Scalling", accuracy_score(y_test,y_predict))
```

# Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
y_predict = dtc.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_predict))
```
